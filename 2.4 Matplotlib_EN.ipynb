### 2.4 MatPlotLib

MatplotLib library is the most widely used library for data visualization. It is intended to generate plots, histograms, spectra, bar charts, error plots, scatterplots, etc., in a few lines of code.

### 2.4.1 How to start Matplotlib?

To use Matplotlib you must import the library **"matplotlib.pyplot "** and its abbreviation is **"plt "**.

```python
import matplotlib.pyplot as plt
```

### 2.4.2 How to generate an X vs Y graph?

One of the most important considerations when generating a plot is that **"the amount of data on each axis of the plot must be the same "** since matplotlib does not work with NaN values. Therefore, it is recommended to check the amount of data on each axis.

### 2.4.2.1 Creation of data to be plotted

$ f(x)=x^x$

$\lim_{x \to 0}{f(x)}$

```python
import numpy as np
X=np.arange(0,11)
Y1=[]
Y2=[]
Y3=[]
Y4=[]
Y5=[]
for x in X:
    a=x
    b=x/10
    c=x/100
    d=x/1000
    e=x/1000000
    y1=a**a
    Y1.append(y1)
    y2=b**b
    Y2.append(y2)
    y3=c**c
    Y3.append(y3)
    y4=d**d
    Y4.append(y4)
    y5=e**e
    Y5.append(y5)
    print(x,y1,y2,y3,y4,y5)
```

```python
#Verificaci√≥n cantidad de datos en cada arreglo/lista
display(X.shape,len(Y1),len(Y2),len(Y3),len(Y4),len(Y5))
```

### 2.4.2.2 "Plot" module  

The "plot" module of the "matplotlib.pyplot" library, or from now on "plt", is the one that starts the generation of the "x vs y" plot. Its main arguments are:
* **scalex, scaley**: share axis between several data.
* **fmt**: vary the color, the "line" style and/or the marker.
* **color**: define the color of the plot. It is called using **c=' '** or **color=' '**.
* **linestyle**: define line style.
* **linewidht**: line thickness.
* **marker**: define marker.
* **markersize**: marker size.
* **label**: define the name to display in the chart legend. It is called using *label=''**; it is activated by adding **"plt.legend() "**.

### 2.4.2.3 Add legends to graph and axes

The **"title() "** command is used to add the title.

The **"xlabel() "** and **"ylabel() "** commands are used to add legends to the axes.

```python
plt.plot(X,Y2,'--m',label='x/10')
plt.plot(X,Y3,'*y',label='x/100')
plt.plot(X,Y4,'r',label='x/1000')
plt.plot(X,Y5,'<b',label='x/1000000')
plt.title('$\lim_{x\to 0}{f(x)}$',fontsize=15)
plt.xlabel('Value of X',fontsize=10)
plt.ylabel('Value of the function')
plt.legend(loc=4);
```

### 2.4.4 Subplots

One way to generate subplots is to use the command **"fig,ax=plt.subplots(rows,columns) "**. This generates an array of axes **ax=[]** and they are called as any array **ax[0:n-1]**. In addition, it has more parameters like sharing the x or y axis ("sharex & sharey") and to define the size of the figure the code **"figsize=(x,y) "** is used.

```python
fig,ax=plt.subplots(2,1,figsize=(7,5))
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral');
```

```python
fig,ax=plt.subplots(1,2,figsize=(8,4))
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
#[0].plot(data.pl_bmassj, data.st_dist)
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral');
```

```python
fig,ax=plt.subplots(1,2,sharey=True)
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral');
```

To adjust the spacing between the plots, **"fig.subplots_adjust(hspace=float,wspace=float) "** is used.

```python
fig,ax=plt.subplots(1,2)
fig.subplots_adjust(wspace=2.0)
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral');
```

```python
fig,ax=plt.subplots(2,1,figsize=(8,4))
fig.subplots_adjust(hspace=2.0)
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral');
```

To add title use **"ax[n].title_set_text('title) "**.

```python
fig,ax=plt.subplots(2,1,figsize=(8,4))
fig.subplots_adjust(hspace=0.7)
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral')
ax[0].title.set_text('Planetary mass referenced to the mass of Jupiter vs distance from Earth')
ax[1].title.set_text('Planetary density vs distance from Earth')
```

```python
fig,ax=plt.subplots(2,1,figsize=(8,4))
fig.subplots_adjust(hspace=1)
ax[0].scatter(data.pl_bmassj, data.st_dist,c=data.pl_bmassj,cmap='Spectral')
ax[1].scatter(data.pl_dens, data.st_dist,c=data.pl_dens,cmap='Spectral')
ax[0].title.set_text('Planetary mass referenced to the mass of Jupiter vs distance from Earth')
ax[1].title.set_text('Planetary density vs distance from Earth')
ax[0].set_xlabel('Distance [pc]')
ax[0].set_ylabel('Planetary mass refer to Jupiter mass')
ax[1].set_xlabel('Distance [pc]')
ax[1].set_ylabel('Planetary Density')
ax[0].set_xticks((0,0.1,0.2,0.3,0.5,1,2,3,4,5,6));
```

### 2.4.5 Pie Chart

```python
labels='Dogs', 'Cats', 'Rabbits', 'Eagles'
sizes=[55,45,9,1]
explode=(0,0,0,0)
fig,ax=plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,shadow=True,startangle=20,autopct='%1.2f%%')
ax.axis('equal')
ax.set_title("% Human Pets ");
```

```python
plt.bar(labels,sizes)
plt.xlabel('Animal')
plt.ylabel('Percentage of domestic presence [%]');
```

```python
c=0
d=0
total=len(data.pl_bmassj)
for pl in data.pl_bmassj:
    if pl>1.0:
        c=c+1
    if pl<0.5:
        d=d+1
        
prc_c=(c/total)*100
prc_d=d/total*100

display(c,d,total,prc_c,prc_d);
```

```python
labels='Greater than Jupiter', 'Less than Jupiter'
sizes=[prc_c,prc_d]
explode=(0.1,0)
fig,ax=plt.subplots()
ax.pie(sizes,explode=explode,labels=labels,shadow=True,startangle=255,autopct='%1.2f%%')
ax.axis('equal')
ax.set_title('Percentage of planetary masses with respect to Jupiter',fontsize=15);
```
