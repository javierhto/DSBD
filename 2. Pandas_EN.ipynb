###  2.2 Pandas 

### 2.2.1 DataFrame

The key structure or object of Pandas is the **DataFrame**. The DataFrame is basically a tabular data structure, with rows and columns. The rows have a specific index to access the data in them and can be identified with names or values; the columns are called **Series**, which essentially consists of a list of values, each one of them with an index.
In synthesis, a DataFrame can be seen as a spreadsheet, but it is much more flexible and the data can be worked as a whole, segmented or individually.

### 2.2.2 Importing Pandas 

To use the Pandas library, the call should be made at the beginning of the document. The standard abbreviation for using Pandas is "**pd**".

```python
import pandas as pd
```

### 2.2.3 DataFrame Creation

A DataFrame can be created mainly from lists, arrays and dictionaries, supporting different formats such as CSV, JSON, H5 and text.

Data can come from different sources. Typically it is obtained from 3 main sources:

* **Personal Hard Drive**- This data is hosted locally and Python manages to access it through paths of this style **c:data/myfile.csv or Users/data/myfile.csv**.

* **Internet** - Data that resides in the cloud, and Python manages to access it through a URL,e.g., https://raw.githubusercontent.com/rsquaredacademy-education/datasets/master/customer.csv.

* **Google Drive(cloud)**- When working with Google CoLab, Google Drive is used to save and load some data. CoLab creates a path to Drive similar to the following syntax: **/content/drive/My Drive/myfile.csv**.

### Iloc & Loc

**ILOC**: method to select elements based on their location.

 * Syntax: **data.iloc[rows,columns]**.
 * Selection of a row/column --> Series.
 * Selection of several rows/columns --> DataFrame.

**LOC** : method to select rows and/or columns based on a label/value or based on a condition. Includes the last term of your indexing.

* Syntax: **data.loc[rows,columns]**.
* Accepts list of labels in columns.
* Conditions can be evaluated to filter data.

```python
data.iloc[30:41]
```

```python
data.loc[30:41]
```

```python
data.loc[30:41, ['pl_bmassj','st_dist']]
```

```python
data[data.loc[0:45:,['pl_bmassj','st_dist']]<150.0]
```

### 2.2.3 Handling Missing Data

Indeterminate or non-existent data is referred to as NaN Numpy works with them through the nan module (np.nan) and Pandas handles them natively within its basic objects. 

```python
import numpy as np
df = pd.DataFrame([np.arange(33,36)],index=['a','b','c'],
columns=["X","Y","Z"]) 
df
```

```python
df=df.reindex(index=['a','b','c','d'])
df
```

#### 2.2.3.1 Isna

Check for missing values

```python
pd.isna(df)
```

####  2.2.3.2 Notna

Check for actual values

```python
pd.notna(df)
```

#### 2.2.3.3 Fillna

### Interpolate

Interpolates NaN values, provided that they belong to a time series, i.e. each value has a specific observation time.

```python
arr=np.array([-10,-5,-3])
arr
```

```python
idx=np.array([pd.Timestamp('20210215'),
pd.Timestamp('20210217'),
pd.Timestamp('20210219')])
idx
```

```python
df = pd.DataFrame(arr,index=idx)
df
```

```python
#Adding intermediate days
idx=[pd.Timestamp('20210215'),
pd.Timestamp('20210216'),
pd.Timestamp('20210217'),
pd.Timestamp('20210218'),
pd.Timestamp('20210219')]
idx
```

```python
df=df.reindex(index=idx)
df
```

```python
df.interpolate().plot()
```

```python
df.interpolate(method="polynomial", order=2).plot()
```
